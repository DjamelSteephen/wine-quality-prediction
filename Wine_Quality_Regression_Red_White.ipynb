# 5-1-Identification des caracteristiques les ples performantes

# Pour le vin rouge

```python
# 1. Obtenir les noms des variables (avec interactions polynomiales)
feature_names_Red= poly_Red.get_feature_names_out(input_features=X_train_Red.columns)

# 2. Associer les coefficients à ces variables
coef_df_Red = pd.DataFrame({
    "Feature": feature_names_Red,
    "Coef": model_poly_Red.coef_
})

# 3. Calcul de l’importance absolue
coef_df_Red["Importance"] = coef_df_Red["Coef"].abs()

# 4. Sélection du top 10 des variables les plus influentes
top_Red = coef_df_Red.sort_values(by="Importance", ascending=False).head(10)

# 5. Construction de la liste de couleurs
colors_Red = top_Red["Coef"].apply(lambda v: "tab:green" if v > 0 else "tab:red").tolist()

# 6. Affichage du barplot avec annotations colorées
plt.figure(figsize=(10, 6))
sns.barplot(
    x="Importance",
    y="Feature",
    data=top_Red,
    color=None,
    orient="h"
)

# Coloration manuelle des barres
for bar, color in zip(plt.gca().patches, colors_Red):
    bar.set_color(color)

# Ajout des valeurs numériques sur les barres
for i, row in enumerate(top_Red.itertuples()):
    sign = "+" if row.Coef >= 0 else "–"
    plt.text(
        row.Importance + 0.01,
        i,
        f"{sign}{abs(row.Coef):.3f}",
        va="center"
    )

plt.title("Top 10 variables polynomiales les plus influentes du vin rouge")
plt.xlabel("Importance (|coef|)")
plt.ylabel("Caractéristique")
plt.tight_layout()
plt.show()
```
